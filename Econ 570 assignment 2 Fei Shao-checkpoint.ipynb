## 1. Covariates


### 1. 1 Simulate a DGP

Simulate a DGP with function $y_i = a+b*T_i+c*x_i+e_i$
With units number $N=100$. $x_i$ represents the covariates

```python
a=1
b=2
c=3
p=0.5
n = 100
X = np.random.normal(5, 2, n)
T = np.random.binomial(1, 0.5, n)
e = np.random.normal(0, 1, n)
Y = a + b*T + c*X + e
print(Y, X, T)
```

### 1.2 DAG

```python
g = gr.Digraph()
g.edge("Treatment (T)", "Outcome(Y)")
g.edge("Covariates (X)", "Outcome(Y)")
g.node("Covariates (X)", color="red")
g
```

### 1.3 Real life example

Take supermarket promotion as a treatment and spending by customer as outcome. When we want to estimate how much the promotion increase spending by customer, it would be affected by other covariates, like the customers' income or the type of product that the customers willing to buy.   
If we interview all the customers and control their income and preferance etc., the estimates would be precise.Otherwise, the error term would be correlated with depending variable and the estimated treatment effect would have bias, like if the chosen experiment group included more low income customers, the treatment effect of the promotion would probably be overestimated.   
But in large samples, this ommited covariates bias' influence would decrease. If you chose enough customers, and the income has a noemal distriution, the probablity of chosing a below average income unit would converge to 50%.

### 1.4 Compare the estimated treatment effect

**For the case that we fail to control for any covariates, use Monte Carlo experiment with sample sizes N=100 and N=1000, to show the bias,RMSE and size of treatment effect estimated:**

```python
class MC_Simulator():
    
    def __init__(self,n_rep):
        """
        n_rep: number of MC replications
        """
        self.n_rep = n_rep
    def simulate_data(self, a, b, c, n,sigma, Control):
        """
        a: intercept
        b: slope of treatment dummy variable
        c: slope of covariate
        n: sample size
        sigma: error variance
        Control: Is variable Xcontroled and included in the estimation for b
        Y = a+b*T+c*X+e
        """
        X = np.random.normal(5, 2, n)
        T = np.random.binomial(1, 0.5, n)
        e = np.random.normal(0, 1, n)
        Y = a + b*T + c*X + e
        
        if Control==1 :
            return (Y,T,X) 
        elif Control==0 :
            return (Y,T)
    
    def estimate_ols_params_ctrl(self,Y,T,X,Control):
        
        mod = sm.OLS(Y, sm.add_constant(pd.concat([pd.Series(T), pd.Series(X)], axis=1)))
        res = mod.fit()
        
        return res.params
    
    def estimate_ols_params_nonctrl(self,Y,T,Control):
        
        mod = sm.OLS(Y, sm.add_constant(T))
        res = mod.fit() 
        
        return res.params
    
    def run_mc_simulation(self,a,b,c,n,sigma, Control):
        
        treatment_effect = []
        bias = []
        RMSE = []
        for rep in range(self.n_rep):
            if Control==1 :
                Y,T,X = self.simulate_data(a, b, c, n, sigma,Control)
                a_hat,b_hat,c_hat = self.estimate_ols_params_ctrl(Y, T, X,Control)
                y_hat = a_hat + b_hat * T + c_hat * X
                treatment_effect = treatment_effect + [b_hat]
                bias = bias + [np.mean(Y[T==1])-np.mean(Y[T==0])]
                RMSE = RMSE + [np.sqrt(np.mean((Y-y_hat)**2))]
            elif Control==0 : 
                Y,T = self.simulate_data(a, b, c, n, sigma,Control)
                a_hat,b_hat = self.estimate_ols_params_nonctrl(Y, T,Control)
                y_hat = a_hat + b_hat * T
                treatment_effect = treatment_effect + [b_hat]
                bias = bias + [np.mean(Y[T==1])-np.mean(Y[T==0])]
                RMSE = RMSE + [np.sqrt(np.mean((Y-y_hat)**2))]
        return bias, RMSE, treatment_effect
    
    def summarize_mc_simulation(self,a,b,c,n,sigma, Control):
        bias, RMSE, treatment_effect = self.run_mc_simulation(a,b,c,n,sigma, Control)
        return [np.mean(bias),np.mean(RMSE),np.mean(treatment_effect)]
```

**For the case that we control for all the covariates that affect the outcome:**

```python
# N=100
mcSim_100 = MC_Simulator(1000)
res_100 = mcSim_100.summarize_mc_simulation(1,2,3,100,1,1)
# N=1000
mcSim_1000 = MC_Simulator(1000)
res_1000 = mcSim_1000.summarize_mc_simulation(1,2,3,1000,1,1)
res = pd.concat([pd.Series(res_100),pd.Series(res_1000)],axis = 1)
res.columns = [['N=100','N=1000']]
res.index = [['bias','RMSE','treatment effect']]
print(res)
```

**For the case that we not control for any covariates that affect the outcome:**

```python
# N=100
mcSim_100 = MC_Simulator(1000)
res_100 = mcSim_100.summarize_mc_simulation(1,2,3,100,1,0)
# N=1000
mcSim_1000 = MC_Simulator(1000)
res_1000 = mcSim_1000.summarize_mc_simulation(1,2,3,1000,1,0)
res = pd.concat([pd.Series(res_100),pd.Series(res_1000)],axis = 1)
res.columns = [['N=100','N=1000']]
res.index = [['bias','RMSE','treatment effect']]
print(res)
```

## 2. Confounding

### 2.1 Simulate a DGP
Simulate a DGP with function $y_i = a+b*T_i+c*x_i+e_i$
With units number $N=100$. $x_i$ represents the confounder that influence both Y and T(by influencing the choosing of unit)

```python
a=1
b=2
c=3
p=1/(1+X**2)
n = 100
T = np.random.binomial(1, p, n)
X = np.random.normal(5,2,n)
e = np.random.normal(0, 1, n)
Y = a + b*T + c*X +e
print(Y, X, T)
```

### 2.3 Real life examples

Take planting tree for example. The outcome is the height of the tree, the treatment is the amount of manure we used, and the confounder is the rainfall. The rainfall would influence both the absorbing of manure and the growth of the tree, so we must control it ot prevent the bias of estimated treatmentt effect.

```python
class MC_Simulator():
    
    def __init__(self,n_rep):
        """
        n_rep: number of MC replications
        """
        self.n_rep = n_rep
    def simulate_data(self, a, b, c, n,sigma, Control):
        """
        a: intercept
        b: slope of treatment dummy variable
        c: slope of the confounder variable
        n: sample size
        sigma: error variance
        Control: Is variable Xcontroled and included in the estimation for b
        Y = a+b*T+c*X+e
        """
        X = np.random.normal(5,2,n)
        T = np.random.binomial(1, 1/(1+X**2), n)
        e = np.random.normal(0, 1, n)
        Y = a + b*T + c*X +e
        
        if Control==1 :
            return (Y,T,X) 
        elif Control==0 :
            return (Y,T)
    
    def estimate_ols_params_ctrl(self,Y,T,X,Control):
        
        mod = sm.OLS(Y, sm.add_constant(pd.concat([pd.Series(T), pd.Series(X)], axis=1)))
        res = mod.fit()
        
        return res.params
    
    def estimate_ols_params_nonctrl(self,Y,T,Control):
        
        mod = sm.OLS(Y, sm.add_constant(T))
        res = mod.fit() 
        
        return res.params
    
    def run_mc_simulation(self,a,b,c,n,sigma, Control):
        
        treatment_effect = []
        bias = []
        RMSE = []
        for rep in range(self.n_rep):
            if Control==1 :
                Y,T,X = self.simulate_data(a, b, c, n, sigma, Control)
                a_hat,b_hat,c_hat = self.estimate_ols_params_ctrl(Y, T, X, Control)
                y_hat = a_hat + b_hat * T + c_hat * X
                treatment_effect = treatment_effect + [b_hat]
                bias = bias + [np.mean(Y[T==1])-np.mean(Y[T==0])]
                RMSE = RMSE + [np.sqrt(np.mean((Y-y_hat)**2))]
            elif Control==0 : 
                Y,T = self.simulate_data(a, b, c, n, sigma, Control)
                a_hat,b_hat = self.estimate_ols_params_nonctrl(Y, T, Control)
                y_hat = a_hat + b_hat * T
                treatment_effect = treatment_effect + [b_hat]
                bias = bias + [np.mean(Y[T==1])-np.mean(Y[T==0])]
                RMSE = RMSE + [np.sqrt(np.mean((Y-y_hat)**2))]
        return bias, RMSE, treatment_effect
    
    def summarize_mc_simulation(self,a,b,c,n,sigma, Control):
        bias, RMSE, treatment_effect = self.run_mc_simulation(a,b,c,n,sigma, Control)
        return [np.mean(bias),np.mean(RMSE),np.mean(treatment_effect)]
```
